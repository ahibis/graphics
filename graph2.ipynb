```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
u=linspace(-20,20,20);
v=linspace(-20,20,20);
x, y = meshgrid(u,v)
z = x**2 + y**2
z1 = (x*y-3)*(x+y)
ax = plt.subplot(121,projection='3d')
ax1 = plt.subplot(122,projection='3d')
ax.plot_surface(x,y,z, cmap='plasma');
ax1.plot_surface(x,y,z1, cmap='viridis');
```

```python
u=linspace(-1.5,1.5,50);
x, y = meshgrid(u,u)
z = x**4 -2*x**2 + y**4 -2*y**2
ax = plt.subplot(121,projection='3d')
ax1 = plt.subplot(122)
ax.plot_surface(x,y,z, cmap='plasma');
cc = ax1.contourf(z, cmap='plasma') 
cbar = plt.colorbar(cc)
```
