```python
from numpy import *
import matplotlib.pyplot as plt
u=arange(-10,10,0.1);
v=arange(-10,10,0.1);
x, y = meshgrid(u,v)
z= sin(x+y*1j).real
ax = plt.subplot(111,projection='3d')
ax.plot_surface(x,y,z, cmap='plasma');

```
