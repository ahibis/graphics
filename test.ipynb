```python
def der(f):
    t=0.001
    def dec(x):
        return (f(x+t)-f(x))/t 
    return dec
```

```python
import matplotlib.pyplot as plt
import numpy as np
from time import *

@der
@der
@der
def func(x):
    return x**10

X = np.arange(-50,50,0.1)
Y = [func(x) for x in X]
plt.plot(X,Y)
    
```
