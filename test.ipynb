```python
def der(f):
    t=0.001
    def dec(x):
        return (f(x+t)-f(x))/t 
    return dec
```

```python
import matplotlib.pyplot as plt
import numpy as np
from time import *

@der
@der
@der
def func(x):
    return x**10

X = np.arange(-50,50,0.1)
Y = [func(x) for x in X]
plt.plot(X,Y)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
X = []
Y = []
v = np.array([1,1])#начальное положение
growth = 1 # уведичение с каждой иттерацией
angle = pi/90 # угол изменения с каждой иттерацией
count = 500 # кол-во иттерацией
rv = np.array([cos(angle),sin(angle)])#вектор поворота на аргумент вектора
mat = growth*np.array([[rv[0],-rv[1]],[rv[1],rv[0]]])#матрица преобразования

v1 = np.array([0.4,0.4])
angle = pi/180
rv = np.array([cos(angle),sin(angle)])#вектор поворота на аргумент вектора
mat1= growth*np.array([[rv[0],-rv[1]],[rv[1],rv[0]]])#матрица преобразования
for i in range(count):
    V=v+v1
    X.append(V[0])
    Y.append(V[1])
    v=mat @ v # перемножаем матрицу преобразования на вектор
    v1=mat1 @ v1
plt.plot(X,Y)
```

```python
import matplotlib.pyplot as plt
from math import *
X = []
Y = []
v = 1+1j#начальное положение
growth = 1 # уведичение с каждой иттерацией
angle = pi/90 # угол изменения с каждой иттерацией
count = 500 # кол-во иттерацией
rot = growth*(cos(angle)+sin(angle)*1j)

for i in range(count):
    X.append(v.real)
    Y.append(v.imag)
    v= v * rot

plt.plot(X,Y)
```
