```python
import matplotlib.pyplot as plt
from numpy import *
def g(x):
    if(x<=0): 
        return 0.5*x+1
    return g(x-1)*2

f = vectorize(g)
x = arange(0,20,0.1);
y = f(x)
y1 = 2**x
plt.plot(x,y)
plt.plot(x,y1)
```

```python
dt = 0.001
R=10
s=-pi*R/2
u=0
t=0
g=10
Tps=[]
Sps=[]
while t<100:
    a=-g*sin(s/(pi*R))
    u+=a*dt
    s+=u*dt
    Tps.append(t)
    Sps.append(s)
    t+=dt
Tps=array(Tps)
#Sinps=-(pi*R/2)*cos(Tps/1.57)
plt.plot(Tps,Sps)
#plt.plot(Tps,Sinps)
p=-1
for i,e in enumerate(Sps):
    if p*e<=0:
        print(i);
    p=e
```
