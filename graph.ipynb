```python
import matplotlib.pyplot as plt
from numpy import *
def g(x):
    if(x<=0): 
        return 0.5*x+1
    return g(x-1)*2

f = vectorize(g)
x = arange(0,20,0.1);
y = f(x)
y1 = 2**x
plt.plot(x,y)
plt.plot(x,y1)
```

```python
x=arange(-50,50);
y=x
z=x**2+y**2
ax = plt.subplot(projection='3d')
ax.plot(x,y,z)
```

```python
u=linspace(-50,50,20);
v=linspace(-50,50,20);
x, y = meshgrid(u,v)
z = x**2 + y**2
z1 = sin(x) + cos(y)
ax = plt.subplot(121,projection='3d')
ax1 = plt.subplot(122,projection='3d')
ax.plot_surface(x,y,z, cmap='plasma');
ax1.plot_surface(x,y,z1, cmap='viridis');
```

```python
x, y = meshgrid(linspace(-50,50,20),linspace(-50,50,20))

u = x / sqrt(x*x + y*y)
v = y/ sqrt(x*x + y*y)
plt.quiver(x,y,u,v)
```
